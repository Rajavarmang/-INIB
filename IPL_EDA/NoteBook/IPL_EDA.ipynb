```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(y='winner', data=matches, order=matches['winner'].value_counts().index)
plt.title('Total Wins by Team')
plt.xlabel('Number of Wins')
plt.ylabel('Team')
plt.show()
```

```python
matches_played = pd.concat([matches['team1'], matches['team2']]).value_counts()
wins = matches['winner'].value_counts()

win_pct = (wins / matches_played * 100).sort_values(ascending=False)
print(win_pct)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=win_pct.values, y=win_pct.index)
plt.title('Win Percentage by Team')
plt.xlabel('Win %')
plt.ylabel('Team')
plt.show()
```

```python
top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
print(top_batsmen)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index)
plt.title('Top 10 Run Scorers in IPL History')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.show()
```

```python
# Check what dismissal types exist first
deliveries['dismissal_kind'].value_counts()
```

```python
bowler_dismissals = ['caught', 'bowled', 'lbw', 'stumped', 'caught and bowled', 'hit wicket']

bowler_wickets = deliveries[deliveries['dismissal_kind'].isin(bowler_dismissals)]

top_bowlers = bowler_wickets.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10)
print(top_bowlers)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index)
plt.title('Top 10 Wicket Takers in IPL History')
plt.xlabel('Total Wickets')
plt.ylabel('Bowler')
plt.show()
```

```python
balls_faced = deliveries.groupby('batsman')['ball'].count()
runs_scored = deliveries.groupby('batsman')['batsman_runs'].sum()

strike_rate = (runs_scored / balls_faced * 100)

# Filter to players with meaningful sample size (e.g., 500+ balls faced) to avoid small-sample noise
qualified_batsmen = balls_faced[balls_faced >= 500].index
top_strike_rate = strike_rate[qualified_batsmen].sort_values(ascending=False).head(10)
print(top_strike_rate)
```

```python
runs_conceded = deliveries.groupby('bowler')['total_runs'].sum()
balls_bowled = deliveries.groupby('bowler')['ball'].count()
overs_bowled = balls_bowled / 6

economy_rate = runs_conceded / overs_bowled

# Filter to bowlers with meaningful sample (e.g., 300+ balls bowled)
qualified_bowlers = balls_bowled[balls_bowled >= 300].index
best_economy = economy_rate[qualified_bowlers].sort_values(ascending=True).head(10)
print(best_economy)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_strike_rate.values, y=top_strike_rate.index)
plt.title('Top 10 Batsmen by Strike Rate (min 500 balls faced)')
plt.xlabel('Strike Rate')
plt.ylabel('Batsman')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=best_economy.values, y=best_economy.index)
plt.title('Top 10 Bowlers by Economy Rate (min 300 balls bowled)')
plt.xlabel('Economy Rate (runs/over)')
plt.ylabel('Bowler')
plt.show()
```
